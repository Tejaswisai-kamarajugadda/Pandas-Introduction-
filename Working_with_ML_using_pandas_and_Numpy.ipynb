```python
# EDA - Exploratory Data Analysis - Matplotlib

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.hist(df['age'], bins = 20, color = 'blue', edgecolor = 'black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age based EDA')
plt.grid(True, axis = 'y', linestyle = '--')
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['income'].value_counts().plot(kind = 'bar',color = ['blue','orange'] ,)
plt.xlabel('Income category')
plt.ylabel('count')
plt.title('Income Distribution')
plt.grid(True, linestyle = '--', axis = 'y')
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.boxplot(column='age',by='income', grid=False)
plt.xlabel('Income category')
plt.ylabel('Age')
plt.title('Age distribution by income')
plt.suptitle("")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
for col in df.select_dtypes(include=['object']).columns:
    df[col] = LabelEncoder().fit_transform(df[col])
```

```python
# Split dataset

X = df.drop(columns=['income'])
y = df['income']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
# Train Linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict and evaluate
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean squared error : {mse}')
```

```python
newPrediction_1 = np.array([[39,2,77777,4,0,13,2020,0,40,39,1,0,1,0]])
```

```python
newPrediction_1 = newPrediction_1.reshape(1,-1)
predictedIncome = model.predict(newPrediction_1)
```
